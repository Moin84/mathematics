```python
import random

X = []
Y = []

for i in range(10):
    x = []
    y = []

    i = 1
    num = random.randint(800, 1000)

    while True:
        y.append(num)
        x.append(i)
        i += 1
        if num == 1:
            break
        if num % 2 == 0:
            num = num / 2
        else:
            num = num * 3 + 1
    X.append(x)
    Y.append(y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))

for i in range(len(X)):
    plt.plot(X[i], Y[i], label=f'Series {i + 1}')

plt.grid(True)
plt.show()
```
